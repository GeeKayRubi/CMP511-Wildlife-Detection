```python
class WildLife(Dataset):
    def __init__(self, root_folder, transform=None):
        self.root_folder = root_folder
        self.transform = transform
        self.data = []

        self.classes = ["buffalo","elephant", "rhino", "zebra"]  # Add all class names
        
        # Loop through each animal folder
        for animal in os.listdir(root_folder):
            animal_folder = os.path.join(root_folder, animal)
            if not os.path.isdir(animal_folder):
                continue
            
            # Get all image files (assume txt files have the same base name)
            image_files = sorted([f for f in os.listdir(animal_folder) if f.endswith('.jpg')])

            for img_file in image_files:
                txt_file = img_file.replace('.jpg', '.txt')
                txt_path = os.path.join(animal_folder, txt_file)
                img_path = os.path.join(animal_folder, img_file)

                if os.path.exists(txt_path) and os.path.exists(img_path):
                    self.data.append((img_path, txt_path, animal))  # Store image, label, and class
                else:
                    print(f"Skipping {img_file}: missing label or image file")
                    
    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_path, label_path, class_name = self.data[idx]
        image = Image.open(img_path).convert("RGB")

        original_width, original_height = image.size  # Get original image dimensions
        resized_width, resized_height = 256, 256  # Target dimensions after resizing

        boxes = []
        labels = []
        
        with open(label_path, "r") as file:
            for line in file:
                values = list(map(float, line.strip().split()))
                class_id = int(values[0])  # Class ID from YOLO file
                cx, cy, bw, bh = values[1:]  # YOLO format
                
                # Convert YOLO format to Pascal VOC (x_min, y_min, x_max, y_max)
                x_min = (cx - bw / 2) * original_width
                y_min = (cy - bh / 2) * original_height
                x_max = (cx + bw / 2) * original_width
                y_max = (cy + bh / 2) * original_height

                # Adjust for resizing
                scale_x = resized_width / original_width
                scale_y = resized_height / original_height
    
                x_min = x_min * scale_x
                y_min = y_min * scale_y
                x_max = x_max * scale_x
                y_max = y_max * scale_y

                
                # Debug: Print the conversion for the first image
                #if idx == 0:
                    #print(f"YOLO format: cx={cx}, cy={cy}, bw={bw}, bh={bh}")
                    #print(f"Original dimensions: width={original_width}, height={original_height}")
                    #print(f"Resized dimensions: width={resized_width}, height={resized_height}")
                    #print(f"Pascal VOC format (adjusted): x_min={x_min}, y_min={y_min}, x_max={x_max}, y_max={y_max}")
                # Ensure bounding box is within image boundaries
                x_min, y_min, x_max, y_max = max(0, x_min), max(0, y_min), min(resized_width, x_max), min(resized_height, y_max)

                # Fix incorrect boxes
                if x_max > x_min and y_max > y_min:
                    boxes.append([x_min, y_min, x_max, y_max])
                    labels.append(class_id + 1)  # Add 1 because Faster R-CNN expects class IDs to start from 1

        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        labels = torch.as_tensor(labels, dtype=torch.int64)
        
        # Ensure at least one valid bbox exists
        if len(boxes) == 0:
            print(f"No valid boxes for {img_path}, skipping image.")
            return self.__getitem__((idx + 1) % len(self.data))  # Load next valid sample

        target = {"boxes": boxes, "labels": labels}
        
        if self.transform:
            image = self.transform(image)

                # Debug prints
        #print(f"Image path: {img_path}")
        #print(f"Annotation path: {label_path}")
        #print(f"Image size: {image.size}")
        #print(f"Target: {target}")

        return image, target  # Return image and Traget data
```

```python
DEVICE = torch.device("cpu")
BATCHSIZE = 4
CLASSES = 5
DIR = os.getcwd()
EPOCHS = 1
N_TRAIN_EXAMPLES = BATCHSIZE * 30
N_VALID_EXAMPLES = BATCHSIZE * 10

def collate_fn(batch):
    batch = [b for b in batch if b is not None]  # Remove None values

    if len(batch) == 0:
        print("Empty batch detected, skipping...")
        return None  # Prevents crashing on an empty batch
    # Debug prints
    #print(f"Batch size: {len(batch)}")
    #for i, b in enumerate(batch):
    #    print(f"Batch item {i}: {b}")
    return tuple(zip(*batch))
        
def get_data():
    transformTrain = transforms.Compose([
        transforms.Resize((256, 256)),
        transforms.GaussianBlur(9, sigma=(0.1, 0.5)),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue= 0.1),
        #transforms.RandomRotation(10),
        #transforms.RandomHorizontalFlip(0.5),
        #transforms.RandomPerspective(distortion_scale=0.2, p=0.5),
        transforms.ToTensor(),
    ])
    
    transformTest = transforms.Compose([
        transforms.ToTensor(),
    ])
    
    dataset_train = WildLife('archive', transform=transformTrain)
    dataset_test = WildLife('test', transform=transformTest)

    train_loader = DataLoader(dataset_train, batch_size=2, shuffle=True, collate_fn=collate_fn)
    valid_loader = DataLoader(dataset_test, batch_size=1, shuffle=False, collate_fn=collate_fn)

    # Debug prints
    #print(f"Number of training samples: {len(dataset_train)}")
    #print(f"Number of validation samples: {len(dataset_test)}")

    return train_loader, valid_loader




def objective(trial):

    num_classes = 5  # 1 class (elephant) + background
    model = ssdlite320_mobilenet_v3_large(num_classes=num_classes, weights_backbone='DEFAULT', trainable_backbone_layers=0)
    
    # Move model to GPU if available
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    
    # Generate the optimizers.
    optimizer_name = trial.suggest_categorical("optimizer", ["Adam", "RMSprop", "SGD"])
    lr = trial.suggest_float("lr", 1e-5, 1e-1, log=True)
    optimizer = getattr(optim, optimizer_name)(model.parameters(), lr=lr)

    # Get the FashionMNIST dataset.
    train_loader, valid_loader = get_data()

    # Training of the model.
    for epoch in range(EPOCHS):
        model.train()
        for batch_idx, (data, target) in enumerate(train_loader):
            # Limiting training data for faster epochs.
            if batch_idx * BATCHSIZE >= N_TRAIN_EXAMPLES:
                break

            # Debug prints
            #print(f"Training batch {batch_idx}:")
            #print(f"Data: {data}")
            #print(f"Target: {target}")
            
            # Unpack the tuple
            data = [d.to(device) for d in data]
            target = [{k: v.to(device) for k, v in t.items()} for t in target]
           
            optimizer.zero_grad()
            loss_dict = model(data, target)
            losses = sum(loss for loss in loss_dict.values())
            losses.backward()
            optimizer.step()

        # Validation of the model.
        model.eval()
        correct = 0
        with torch.no_grad():
            for batch_idx, (data, target) in enumerate(valid_loader):
                # Limiting validation data.
                if batch_idx * BATCHSIZE >= N_VALID_EXAMPLES:
                    break
                                # Debug prints
                #print(f"Validation batch {batch_idx}:")
                #print(f"Data: {data}")
                #print(f"Target: {target}")
                
                # Unpack the tuple
                data = [d.to(device) for d in data]
                target = [{k: v.to(device) for k, v in t.items()} for t in target]
                
                output = model(data)
                # Extract the predicted labels
                pred_labels = [o['labels'].cpu()  for o in output]
                true_labels = [t['labels'].cpu()  for t in target]

                
                # Debug prints

                #print(f"Validation batch {batch_idx}:")
                #print(f"Predicted labels: {pred_labels}")
                #print(f"True labels: {true_labels}")

                # Calculate the number of correct predictions
                for pred, true in zip(pred_labels, true_labels):
                    if pred.size(0) == 0 or true.size(0) == 0:
                        # Skip comparison if there are no predicted labels
                        continue
                    #print(f"Pred size: {pred.size()}, True size: {true.size()}")
                    min_size = min(pred.size(0), true.size(0))
                    correct += (pred[:min_size] == true[:min_size]).sum().item()

        accuracy = correct / min(len(valid_loader.dataset), N_VALID_EXAMPLES)
        trial.report(accuracy, epoch)

        # Handle pruning based on the intermediate value.
        if trial.should_prune():
            raise optuna.exceptions.TrialPruned()

    return accuracy
```

```python
def calculate_iou(box1, box2):
    """Compute IoU between two bounding boxes."""
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    intersection = max(0, x2 - x1) * max(0, y2 - y1)
    area_box1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area_box2 = (box2[2] - box2[0]) * (box2[3] - box2[1])
    union = area_box1 + area_box2 - intersection

    return intersection / union if union > 0 else 0


def average_precision(precision, recall):
    """Compute AP using Precision-Recall curve."""
    recall = np.concatenate(([0.], recall, [1.]))
    precision = np.concatenate(([0.], precision, [0.]))

    for i in range(len(precision) - 1, 0, -1):
        precision[i - 1] = max(precision[i - 1], precision[i])

    indices = np.where(recall[1:] != recall[:-1])[0]
    ap = np.sum((recall[indices + 1] - recall[indices]) * precision[indices + 1])
    return ap

def mean_average_precision(predictions, ground_truths, iou_threshold=0.5):
    """Compute mAP over multiple images."""
    all_ap = []

    for img_id in ground_truths.keys():
        gt_boxes = ground_truths[img_id]
        pred_boxes, scores = predictions.get(img_id, ([], []))

        sorted_indices = np.argsort(scores)[::-1]
        pred_boxes = [pred_boxes[i] for i in sorted_indices]

        precision, recall, _ = calculate_precision_recall(pred_boxes, gt_boxes, iou_threshold)
        ap = average_precision([precision], [recall])
        all_ap.append(ap)

    return np.mean(all_ap)
```

```python
import os
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import torchvision
import torch.optim as optim
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import numpy as np
from collections import defaultdict
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt


class WildLife(Dataset):
    def __init__(self, root_folder, transform=None):
        self.root_folder = root_folder
        self.transform = transform
        self.data = []

        self.classes = ["buffalo","elephant", "rhino", "zebra"]  # Add all class names
        
        # Loop through each animal folder
        for animal in os.listdir(root_folder):
            animal_folder = os.path.join(root_folder, animal)
            if not os.path.isdir(animal_folder):
                continue
            
            # Get all image files (assume txt files have the same base name)
            image_files = sorted([f for f in os.listdir(animal_folder) if f.endswith('.jpg')])

            for img_file in image_files:
                txt_file = img_file.replace('.jpg', '.txt')
                txt_path = os.path.join(animal_folder, txt_file)
                img_path = os.path.join(animal_folder, img_file)

                if os.path.exists(txt_path) and os.path.exists(img_path):
                    self.data.append((img_path, txt_path, animal))  # Store image, label, and class
                else:
                    print(f"Skipping {img_file}: missing label or image file")
                    
    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_path, label_path, class_name = self.data[idx]
        image = Image.open(img_path).convert("RGB")

        original_width, original_height = image.size  # Get original image dimensions
        resized_width, resized_height = 256, 256  # Target dimensions after resizing

        boxes = []
        labels = []
        
        with open(label_path, "r") as file:
            for line in file:
                values = list(map(float, line.strip().split()))
                class_id = int(values[0])  # Class ID from YOLO file
                cx, cy, bw, bh = values[1:]  # YOLO format
                
                # Convert YOLO format to Pascal VOC (x_min, y_min, x_max, y_max)
                x_min = (cx - bw / 2) * original_width
                y_min = (cy - bh / 2) * original_height
                x_max = (cx + bw / 2) * original_width
                y_max = (cy + bh / 2) * original_height

                # Adjust for resizing
                scale_x = resized_width / original_width
                scale_y = resized_height / original_height
    
                x_min = x_min * scale_x
                y_min = y_min * scale_y
                x_max = x_max * scale_x
                y_max = y_max * scale_y

                
                # Debug: Print the conversion for the first image
                #if idx == 0:
                    #print(f"YOLO format: cx={cx}, cy={cy}, bw={bw}, bh={bh}")
                    #print(f"Original dimensions: width={original_width}, height={original_height}")
                    #print(f"Resized dimensions: width={resized_width}, height={resized_height}")
                    #print(f"Pascal VOC format (adjusted): x_min={x_min}, y_min={y_min}, x_max={x_max}, y_max={y_max}")
                # Ensure bounding box is within image boundaries
                x_min, y_min, x_max, y_max = max(0, x_min), max(0, y_min), min(resized_width, x_max), min(resized_height, y_max)

                # Fix incorrect boxes
                if x_max > x_min and y_max > y_min:
                    boxes.append([x_min, y_min, x_max, y_max])
                    labels.append(class_id + 1)  # Add 1 because Faster R-CNN expects class IDs to start from 1

        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        labels = torch.as_tensor(labels, dtype=torch.int64)
        
        # Ensure at least one valid bbox exists
        if len(boxes) == 0:
            print(f"No valid boxes for {img_path}, skipping image.")
            return self.__getitem__((idx + 1) % len(self.data))  # Load next valid sample

        target = {"boxes": boxes, "labels": labels}
        
        if self.transform:
            image = self.transform(image)

                # Debug prints
        #print(f"Image path: {img_path}")
        #print(f"Annotation path: {label_path}")
        #print(f"Image size: {image.size}")
        #print(f"Target: {target}")

        return image, target  # Return image and Traget data
# Define transformations
transformTrain = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.GaussianBlur(9, sigma=(0.1, 0.5)),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue= 0.1),
    #transforms.RandomRotation(10),
    #transforms.RandomHorizontalFlip(0.5),
    #transforms.RandomPerspective(distortion_scale=0.2, p=0.5),
    transforms.ToTensor(),
])

transformTest = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
])

# Create dataset and dataloader for both training and testing
dataset_train = WildLife('archive', transform=transformTrain)
print(len(dataset_train))
#print(dataset_train)

#for i in range(len(dataset_train)):  # Check first 5 samples
#    img, target = dataset_train[i]
#    print(f"Image {i}: {img.shape}, Boxes: {target['boxes']}")
    
dataset_test = WildLife('test', transform=transformTest)
#print(len(dataset_test))
#print(dataset_test)


# Custom collate function for variable-length bounding boxes
def collate_fn(batch):
    batch = [b for b in batch if b is not None]  # Remove None values

    if len(batch) == 0:
        print("Empty batch detected, skipping...")
        return None  # Prevents crashing on an empty batch

        # Debug prints
    #print(f"Batch size: {len(batch)}")
    #for i, b in enumerate(batch):
    #    print(f"Batch item {i}: {b}")

    return tuple(zip(*batch))

dataloader_train = DataLoader(dataset_train, batch_size=4, shuffle=True, collate_fn=collate_fn)
dataloader_test = DataLoader(dataset_test, batch_size=1, shuffle=False, collate_fn=collate_fn)
print(len(dataloader_train))


#for images, targets in dataloader_train:
   # print(images[0].shape)  # (3, 256, 256)
   # print(targets[0])  # {'boxes': tensor([...]), 'labels': tensor([...])}


# Visualize ground truth bounding boxes
for i in range(3):  # Visualize first 3 images
    img, target = dataset_train[i]
    img_np = img.permute(1, 2, 0).numpy() * 255
    img_np = img_np.astype(np.uint8)

    # Convert RGB to BGR for OpenCV
    img_np = cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR)

    for box in target['boxes']:
        x_min, y_min, x_max, y_max = map(int, box.tolist())
        cv2.rectangle(img_np, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

    # Convert BGR back to RGB for displaying with Matplotlib
    img_np = cv2.cvtColor(img_np, cv2.COLOR_BGR2RGB)

    plt.imshow(img_np)
    plt.axis("off")
    plt.title(f"Ground Truth for Image {i}")
    plt.show()
# Load pre-trained Faster R-CNN model
num_classes = 5  # 1 class (elephant) + background
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True)

# Modify the classifier for our custom dataset
in_features = model.roi_heads.box_predictor.cls_score.in_features
model.roi_heads.box_predictor = torchvision.models.detection.faster_rcnn.FastRCNNPredictor(in_features, num_classes)

# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Define optimizer
optimizer = optim.RMSprop(model.parameters(), lr=4.89974756433587e-05, momentum=0.5, weight_decay=0.0005)

# Define learning rate scheduler
lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.1)


# Initialize lists to store losses
train_losses = []
val_losses = []

# Training loop
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    x=0
    for batch_idx, (images, targets) in enumerate(dataloader_train):
        images = [img.to(device) for img in images]
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

        if all(len(t["boxes"]) == 0 for t in targets):
            continue 

        optimizer.zero_grad()
        loss_dict = model(images, targets)
        losses = sum(loss for loss in loss_dict.values())

        # Skip NaN losses
        if torch.isnan(losses):
            print("NaN detected, skipping batch")
            continue

        losses.backward()
        optimizer.step()
        #print(x)
        epoch_loss += losses.item()
        x=x+1
    
    lr_scheduler.step()
    #print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss}")


    # Average the training loss for the epoch
    epoch_loss /= len(dataloader_train)
    train_losses.append(epoch_loss)
    
# Validation Phase
    model.eval()
    val_loss = 0 
    print("Starting validation...")
    with torch.no_grad():
        for batch_idx, (images, targets) in enumerate(dataloader_test):
            # Move images and targets to the device
            images = [img.to(device) for img in images]
            targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

            model.train()           
            Val_loss_dict = model(images, targets)
            losses = sum(loss for loss in Val_loss_dict.values())
            val_loss += losses.item()
            model.eval()
      

    # Average the validation loss for the epoch
    val_loss /= len(dataloader_test)
    val_losses.append(val_loss)

    #print losses for current epoch
    print(f"Epoch [{epoch+1}/{num_epochs}]")
    print(f"Training Loss:{epoch_loss:.4f}")
    print(f"validation Loss:{val_loss:.4f}")

            # Debug: Print predictions and targets
            #print(f"Validation Batch {batch_idx + 1}:")
            #print(f"Predictions: {predictions}")
            #print(f"Targets: {targets}")

print("Training and Validation Complete!")


# Plot the training and validation loss
plt.figure(figsize=(10, 6))
plt.plot(range(1, num_epochs + 1), train_losses, label="Training Loss", marker='o')
plt.plot(range(1, num_epochs + 1), val_losses, label="Validation Loss", marker='o')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")
plt.legend()
plt.grid()
plt.savefig('Single_training_loss_plot.png')
plt.show()

# Load a test image from dataset
test_img, _ = dataset_test[0]
test_img = test_img.to(device).unsqueeze(0)

# Set model to evaluation mode
model.eval()
with torch.no_grad():
    predictions = model(test_img)

for i, box in enumerate(predictions[0]['boxes']):
    score = predictions[0]['scores'][i].item()
    print(f"Box: {box}, score: {score}")

#print(predictions)
# Convert image to NumPy format for display
img_np = test_img.squeeze(0).cpu().permute(1, 2, 0).numpy()
img_np = (img_np * 255).astype(np.uint8)

#print(img_np.shape)

# Convert RGB to BGR for OpenCV
img_np = cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR)

# Draw predictions
for i, box in enumerate(predictions[0]['boxes']):
    score = predictions[0]['scores'][i].item()
    #print(box)
    #print(score)
    if score > 0.5:  # Confidence threshold
        print('higher than 0.5 con')
        x_min, y_min, x_max, y_max = map(int, box.tolist())
        print(x_min, y_min, x_max, y_max)
        startpoint = (int(x_min),int(y_min))

        endpoint = (int(x_max), int(y_max))
        cv2.rectangle(img_np, startpoint, endpoint,  (0, 255, 0), 2)

# Convert BGR back to RGB for displaying with Matplotlib
img_np = cv2.cvtColor(img_np, cv2.COLOR_BGR2RGB)

# Show image
plt.imshow(img_np)
plt.axis("off")
plt.savefig('model prediction test.png')
plt.show()

def calculate_iou(box1, box2):
    """Compute IoU between two bounding boxes."""
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    intersection = max(0, x2 - x1) * max(0, y2 - y1)
    area_box1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area_box2 = (box2[2] - box2[0]) * (box2[3] - box2[1])
    union = area_box1 + area_box2 - intersection

    return intersection / union if union > 0 else 0


def average_precision(precision, recall):
    """Compute AP using Precision-Recall curve."""
    recall = np.concatenate(([0.], recall, [1.]))
    precision = np.concatenate(([0.], precision, [0.]))

    for i in range(len(precision) - 1, 0, -1):
        precision[i - 1] = max(precision[i - 1], precision[i])

    indices = np.where(recall[1:] != recall[:-1])[0]
    ap = np.sum((recall[indices + 1] - recall[indices]) * precision[indices + 1])
    return ap

def mean_average_precision(predictions, ground_truths, iou_threshold=0.5):
    """Compute mAP over multiple images."""
    all_ap = []

    for img_id in ground_truths.keys():
        gt_boxes = ground_truths[img_id]
        pred_boxes, scores = predictions.get(img_id, ([], []))

        sorted_indices = np.argsort(scores)[::-1]
        pred_boxes = [pred_boxes[i] for i in sorted_indices]

        precision, recall, _ = calculate_precision_recall(pred_boxes, gt_boxes, iou_threshold)
        ap = average_precision([precision], [recall])
        all_ap.append(ap)

    return np.mean(all_ap)

#start evaluation
# Move model to evaluation mode
model.eval()

# Data structures for storing results
all_true_labels = []
all_pred_labels = []
predictions_dict = {}
ground_truths_dict = []

iou_threshold = 0.9  # IoU threshold for TP/FP classification

# Run inference and collect predictions
with torch.no_grad():
    for images, targets in dataloader_test:
        images = list(img.to(device) for img in images)
        outputs = model(images)

        for i, output in enumerate(outputs):
            gt_boxes = targets[i]["boxes"].tolist()
            gt_labels = targets[i]["labels"].tolist()

            pred_boxes = output["boxes"].tolist()
            pred_labels = output["labels"].tolist()
            pred_scores = output["scores"].tolist()

            # Store results for mAP computation
            predictions_dict[i] = (pred_boxes, pred_scores)
            ground_truths_dict.append(gt_boxes)

            # Debug: Print lengths of ground truth and predictions
            print(f"GT Labels: {len(gt_labels)}, Pred Labels: {len(pred_labels)}")

            # Convert ground truth labels
            all_true_labels.extend(gt_labels)

            # Assign predicted labels based on IoU
            for gt_box, gt_label in zip(gt_boxes, gt_labels):
                best_iou = 0
                matched_pred_label = 0  # Background
                for pred_box, pred_label in zip(pred_boxes, pred_labels):
                    iou = calculate_iou(pred_box, gt_box)
                    if iou > best_iou and iou > iou_threshold:
                        best_iou = iou
                        matched_pred_label = pred_label  # Assign predicted class

                all_true_labels.append(gt_label)
                all_pred_labels.append(matched_pred_label)

            # Handle unmatched predictions (false positives)
            for pred_box, pred_label in zip(pred_boxes, pred_labels):
                best_iou = 0
                for gt_box in gt_boxes:
                    iou = calculate_iou(pred_box, gt_box)
                    if iou > best_iou:
                        best_iou = iou

                if best_iou < iou_threshold:  # If no match, it's a false positive
                    all_true_labels.append(0)  # Background
                    all_pred_labels.append(pred_label)

        # Debug: Print lengths after processing each batch
        print(f"Length of all_true_labels: {len(all_true_labels)}")
        print(f"Length of all_pred_labels: {len(all_pred_labels)}")

# Compute Precision, Recall, and F1-score
precision = precision_score(all_true_labels, all_pred_labels, average='weighted', zero_division=1)
recall = recall_score(all_true_labels, all_pred_labels, average='weighted', zero_division=1)
f1 = f1_score(all_true_labels, all_pred_labels, average='weighted')

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1:}")

# Compute mAP
map_score = mean_average_precision(predictions_dict, ground_truths_dict)
print(f"mAP: {map_score:}")

# Generate confusion matrix
cm = confusion_matrix(all_true_labels, all_pred_labels)

# Plot using seaborn
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Buffalo", "Elephant","Rhino", "Zebra", "Background"], yticklabels=["Buffalo", "Elephant","Rhino", "Zebra", "Background"])
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Confusion Matrix")
plt.show()
```
